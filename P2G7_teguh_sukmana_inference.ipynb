```python
# Import libraries

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
# Load the trained model

model_path = "/content/drive/MyDrive/Hacktiv8/CNN_smoking_model.keras"
model = load_model(model_path)
```

```python
# Define function to preprocess input image
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(250, 250)) # load the image
    img_array = image.img_to_array(img) # convert to array
    img_array = np.expand_dims(img_array, axis=0) / 255.0  # Normalize pixel values
    return img_array
```

```python
# Define function to make predictions
def predict_image(img_path):
    img_array = preprocess_image(img_path)
    prediction = model.predict(img_array)
    return prediction[0][0] # probablity
```

```python
# Load an example image for testing 1
test_img_path_1 = "/content/drive/MyDrive/Hacktiv8/smoking.jpg"
test_img_1 = image.load_img(test_img_path_1)
plt.imshow(test_img_1)
plt.axis('off')
plt.show()
```

```python
# Load an example image for testing 2
test_img_path_2 = "/content/drive/MyDrive/Hacktiv8/notsmoking.jpg"
test_img_2 = image.load_img(test_img_path_2)
plt.imshow(test_img_2)
plt.axis('off')
plt.show()
```

```python
# Perform prediction 1
prediction = predict_image(test_img_path_1)
if prediction >= 0.5:
    print("Predicted: Smoking Image")
else:
    print("Predicted: Not-Smoking Image")
```

```python
# Perform prediction 2
prediction = predict_image(test_img_path_2)
if prediction >= 0.5:
    print("Predicted: Smoking Image")
else:
    print("Predicted: Not-Smoking Image")
```
